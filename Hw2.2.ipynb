```python
# Task 6 Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
dfn = df.drop(["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1)
```

```python
# Task 10 Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python   
python_series = dfn["Язык.программирования"] == "Python"
python_data = dfn.loc[python_series] 
python_data
```

```python
# Task 11 Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
# Task 12  Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
dev_df_gr = python_data.groupby("Должность")
dev_df_gr
```

```python
# Task 13 Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
min_max_sal = dev_df_gr["Зарплата.в.месяц"].agg(["min", "max"])
```

```python
min_max_sal
```

```python
# Task 14 Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    row["avg"] = row[["min", "max"]].mean()
    row["position"] = row.name
    return row


df_new = min_max_sal.apply(fill_avg_salary, axis=1)
df_new = df_new[["position", "min", "max", "avg"]]
```

```python
df_new
```

```python
# Task 15 Створіть описову статистику за допомогою методу describe для нового стовпчика.
df_new["avg"].describe()
```

```python
# Task 16 Збережіть отриману таблицю в CSV файл
df_new.to_csv("python_salary.csv", index=False)
```
